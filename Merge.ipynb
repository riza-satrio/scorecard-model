```python
counts_prev = prev.groupby('SK_ID_CURR').size().reset_index(name='COUNT_PREVIOUS_APPLICATIONS')

counts_prev.head(10)
```

```python
counts_bureau = bureau.groupby('SK_ID_CURR').size().reset_index(name='COUNT_PREVIOUS_OTHER')

counts_bureau.head(10)
```

```python
bureau_balance = pd.read_csv("bureau_balance.csv")
bureau_balance["STATUS"].value_counts()
```

```python
# Merging counts_prev with df
df_merged = df.merge(counts_prev, on='SK_ID_CURR', how='left')

# Merging counts_bureau with the intermediate df_merged
df_merged = df_merged.merge(counts_bureau, on='SK_ID_CURR', how='left')

df_merged.head(20)
```
